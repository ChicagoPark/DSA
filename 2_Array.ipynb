## [3] Finding an element 

```python
arr1 = array('i', [1,2,3,4,5,6])

def searchInArray(array, value):
    for i in arr1:
        if i == value:
            return arr1.index(value) # index method get the value and then return the index
    return "The element does not exist in this array" 

print(searchInArray(arr1, 5))
    
```

# Two Dimensional Array

#### When we create an array, we:
* Assign it to a variable
* Define the type of elements that it will store
* Define its size (the maximum numbers of elements)

```python
import numpy as np

twoDArray = np.array([[11,15,10,6],[10,14,11,5],[12,17,12,8],[15,18,14,9]])
print(twoDArray)
```

## [1] Insertation - Two Dimensional Array

```python
twoDArray = np.array([[11,15,10,6],[10,14,11,5],[12,17,12,8],[15,18,14,9]])

# Add new column to twoDArray
newTwoDArray = np.insert(twoDArray, 0, [[1,2,3,4]], axis=0)

print(newTwoDArray)

# Realization Note: In the list case, we can append value to the list with append method without putting variable number at the left side.
                # But, when we use append method not in the list but in np, we need to save it into the array.

newTwoDArray = np.append(newTwoDArray, [[1,2,3,4]], axis=0)

print(newTwoDArray)
```

## [2] Accessing

```python
def accessElements(array, rowIndex, colIndex):
    assert array.shape != (0,0) or rowIndex >= 0 or rowIndex < array.shape[0] or colIndex >= 0 or colIndex < array.shape[1], 'Put the proper value'
    print(array[rowIndex, colIndex])
print(newTwoDArray.shape[0])
```

## [3] Traverse

```python
def traverse(array):
    assert array.shape != (0,0), 'Put the array'
    for i in range(0, len(array)):
        for j in range(0, len(array[1])):
            print(array[i][j], end=' ')
        print('')
            
traverse(newTwoDArray)
```

## [4] Deletion - Two Dimensional Array

```python
twoDArray = np.array([[11,15,10,6],[10,14,11,5],[12,17,12,8],[15,18,14,9]])
print(twoDArray)

newTDArray = np.delete(twoDArray, 1, axis = 0)

print(newTDArray)
```
